```python
import math

epsilon = 1e-3

sum_series = 0
n = 0
a_n = 1

while a_n >= epsilon:
    sum_series += a_n
    n += 1
    a_n = (10 ** n) / math.factorial(n)

print("Summa ryada s epsilon = 10^-3 = ", sum_series)
```

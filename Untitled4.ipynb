```python
class Document:
  def __init__(self, name, number):
    self.name = name
    self.number = number

class Company(Document):
  def __init__(self, name, number, address):
    super().__init__(name, number)
    self.address = address

  def update_address(self, newname):
    self.address = newname

  def display_info(self):
    print(f'Имя компании: {self.name}, Номер компании: {self.number}, Адрес компании: {self.address}')

x = Company('BBC', 313, 'Kazan')
x.update_address('Moscow')
x.display_info()
```

```python
class Company:
  def __init__(self, companyName, address):
    self.__companyName = companyName
    self.__address = address
  @property
  def address(self):
    return self.__address

  @address.setter
  def address(self, newaddress):
    self.__address = newaddress

  @property
  def companyName(self):
    return self.__companyName

  @companyName.setter
  def companyName(self, newname):
    self.__companyName = newname


  def display_info(self):
    print(f'ИМя компании {self.companyName}, Адрес компании {self.address}')


x = Company("BBC", "Moscow")
x.address = "Kazan"
x.companyName = "KFU"
x.display_info()
```

5. Абстрактные классы

image.png

```python
from abc import ABC, abstractmethod
class Function(ABC):
  @abstractmethod
  def evaluate(self, x):
    pass

class LinearFunction(Function):

  def __init__(self, a, b):
    self.a = a
    self.b = b

  def evaluate(self, x):
    return self.a*x + self.b
class QuadraticFunction(Function):

  def __init__(self, a, b, c):
    self.a = a
    self.b = b
    self.c = c

  def evaluate(self, x):
    return self.a*x*x+self.b*x+self.c

linear_func = LinearFunction(2, 3)  # y = 2x + 3
quadratic_func = QuadraticFunction(1, -2, 1) # y = x^2 - 2x + 1

print(f"Linear function x=5: {linear_func.evaluate(5)}")
print(f"Quadratic function x=3: {quadratic_func.evaluate(3)}")
```

```python
from abc import ABC, abstractclassmethod

class Company(ABC):

  @abstractmethod
  def generateReport(self):
    pass

class TechCompany(Company):
  def __init__(self, name, profit, successful_transactions):
    self.name = name
    self.profit = profit
    self.successful_transactions = successful_transactions

  def generateReport(self):
    print(f'Имя компании: {self.name}, Выручка компании {self.profit}$, Успешных сделок: {self.successful_transactions}')

class ManufacturingCompany(Company):
  def __init__(self, name, profit, successful_transactions):
    self.name = name
    self.profit = profit
    self.successful_transactions = successful_transactions

  def generateReport(self):
    print(f'Имя компании: {self.name}, Выручка компании {self.profit}$, Успешных сделок: {self.successful_transactions}')

TC = TechCompany("Nvidia", 1000, 300)
MC = ManufacturingCompany("Boom", 2000, 500)

TC.generateReport()
MC.generateReport()
```

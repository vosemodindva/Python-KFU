8. Процедуры и функции: группа Proc


```python

def odd(x):
    while x > 0:
        digit = x % 10
        if digit % 2 == 0:
            return False
        x = x // 10
    return True

def sum_odd_num():
    s = 0
    for i in range(100, 1000):
        if odd(i) == True:
            s += i
    return s

def even_num_insum(x):
    k = 0
    while (x > 0):
        n = x % 10
        x = x // 10
        if (n % 2 == 0):
            k+=1
    return k


print(sum_odd_num())
print(even_num_insum(sum_odd_num()))
```

```python

m = int(input())
n = int(input())

def fac(x):
    s = 1
    for i in range(1, x + 1):
        s = s * i
    return s

def sum_fac():
    s = 0
    for i in range(m, n + 1):
        if (i % 2 == 0):
            s = s + fac(i)
    print(s)

sum_fac()
```

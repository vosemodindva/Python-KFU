```python
import json
import logging
from datetime import datetime
from abc import ABC, ABCMeta, abstractmethod
from functools import wraps

# Настройка логов
logging.basicConfig(filename='rental.log', level=logging.INFO)

# Ошибки
class PropertyError(Exception):
    pass

class PermissionError(Exception):
    pass

class AgreementError(Exception):
    pass

# Интерфейсы
class Rentable(ABC):
    @abstractmethod
    def rent_property(self, tenant, start, end):
        pass

class Reportable(ABC):
    @abstractmethod
    def report(self):
        pass

# Миксины
class LoggingMixin:
    def log(self, message):
        logging.info(message)

class NotificationMixin:
    def notify(self, message):
        print(f"Уведомление: {message}")

# Метакласс для Property
class PropertyMeta(ABCMeta):
    registry = {}

    def __new__(cls, name, bases, attrs):
        new_cls = super().__new__(cls, name, bases, attrs)
        if name != 'Property':
            cls.registry[name.lower()] = new_cls
        return new_cls

# Базовый класс для недвижимости
class Property(Rentable, Reportable, LoggingMixin, NotificationMixin, metaclass=PropertyMeta):
    def __init__(self, id, address, area, rate):
        if area <= 0 or rate < 0:
            raise PropertyError("Площадь или ставка неверные")
        self._id = id
        self._address = address
        self._area = area
        self._rate = rate
        self._available = True
        self.log(f"Создана недвижимость {id} по адресу {address}")
        self.notify(f"Недвижимость {address} добавлена")

    def get_id(self):
        return self._id

    def get_address(self):
        return self._address

    def set_address(self, address):
        self._address = address
        self.log(f"Адрес изменен для {self._id}")

    def get_area(self):
        return self._area

    def get_rate(self):
        return self._rate

    def set_rate(self, rate):
        if rate < 0:
            raise PropertyError("Ставка не может быть отрицательной")
        self._rate = rate
        self.log(f"Ставка изменена для {self._id}")

    def is_available(self):
        return self._available

    def set_available(self, status):
        self._available = status
        self.log(f"Статус недвижимости {self._id}: {'доступна' if status else 'занята'}")

    def calculate_cost(self, months):
        pass

    def __str__(self):
        return f"Недвижимость: {self._address}, Площадь: {self._area} кв.м"

    def __lt__(self, other):
        return self._rate < other.get_rate()

    def __gt__(self, other):
        return self._rate > other.get_rate()

    def __eq__(self, other):
        return self._id == other.get_id()

    def report(self):
        return f"ID: {self._id}, Адрес: {self._address}, Площадь: {self._area}, Ставка: {self._rate}, Доступна: {self._available}"

    def rent_property(self, tenant, start, end):
        if not self._available:
            raise PropertyError(f"Недвижимость {self._address} занята")
        self._available = False
        agreement = RentalAgreement(f"agr_{self._id}", tenant, self, start, end)
        self.log(f"Недвижимость {self._id} арендована")
        self.notify(f"Аренда подтверждена для {self._address}")
        return agreement

    def to_dict(self):
        return {
            "id": self._id,
            "address": self._address,
            "area": self._area,
            "rate": self._rate,
            "available": self._available,
            "type": self.__class__.__name__.lower()
        }

    @staticmethod
    def from_dict(data):
        prop_type = data["type"]
        prop_class = PropertyMeta.registry.get(prop_type)
        if not prop_class:
            raise PropertyError(f"Неверный тип: {prop_type}")
        return prop_class(data["id"], data["address"], data["area"], data["rate"], **data.get("extra", {}))

# Классы для недвижимости
class Apartment(Property):
    def __init__(self, id, address, area, rate, rooms):
        super().__init__(id, address, area, rate)
        self._rooms = rooms

    def calculate_cost(self, months):
        cost = self._rate * months
        if months > 12:
            cost *= 0.9
        return cost

    def __str__(self):
        return f"Квартира: {self._address}, Комнат: {self._rooms}, Площадь: {self._area} кв.м"

    def to_dict(self):
        data = super().to_dict()
        data["extra"] = {"rooms": self._rooms}
        return data

class House(Property):
    def __init__(self, id, address, area, rate, garden):
        super().__init__(id, address, area, rate)
        self._garden = garden

    def calculate_cost(self, months):
        cost = self._rate * months
        if self._garden:
            cost += 200 * months
        return cost

    def __str__(self):
        return f"Дом: {self._address}, Сад: {'есть' if self._garden else 'нет'}, Площадь: {self._area} кв.м"

    def to_dict(self):
        data = super().to_dict()
        data["extra"] = {"garden": self._garden}
        return data

class Commercial(Property):
    def __init__(self, id, address, area, rate, business):
        super().__init__(id, address, area, rate)
        self._business = business

    def calculate_cost(self, months):
        cost = self._rate * months
        if self._business.lower() == "retail":
            cost *= 1.15
        return cost

    def __str__(self):
        return f"Коммерция: {self._address}, Бизнес: {self._business}, Площадь: {self._area} кв.м"

    def to_dict(self):
        data = super().to_dict()
        data["extra"] = {"business": self._business}
        return data

# Класс арендатора
class Tenant:
    def __init__(self, id, name, contact):
        self._id = id
        self._name = name
        self._contact = contact

    def get_id(self):
        return self._id

    def get_name(self):
        return self._name

    def to_dict(self):
        return {"id": self._id, "name": self._name, "contact": self._contact}

    @staticmethod
    def from_dict(data):
        return Tenant(data["id"], data["name"], data["contact"])

# Класс договора аренды
class RentalAgreement(LoggingMixin, NotificationMixin, Reportable):
    def __init__(self, id, tenant, property, start, end):
        self._id = id
        self._tenant = tenant
        self._property = property
        self._start = start
        self._end = end
        self._extras = {}
        self._total_cost = self.calculate_total()
        self.log(f"Создан договор {id}")
        self.notify(f"Аренда подтверждена для {property.get_address()}")

    def calculate_total(self):
        months = (self._end - self._start).days / 30
        cost = self._property.calculate_cost(months)
        for extra_cost in self._extras.values():
            cost += extra_cost
        return cost

    def add_extra(self, service, cost):
        self._extras[service] = cost
        self._total_cost = self.calculate_total()
        self.log(f"Добавлена услуга {service} в договор {self._id}")

    def remove_extra(self, service):
        if service in self._extras:
            del self._extras[service]
            self._total_cost = self.calculate_total()
            self.log(f"Удалена услуга {service} из договора {self._id}")

    def report(self):
        return f"Договор: {self._id}, Арендатор: {self._tenant.get_name()}, Недвижимость: {self._property.get_address()}, Стоимость: {self._total_cost}"

    def to_dict(self):
        return {
            "id": self._id,
            "tenant": self._tenant.to_dict(),
            "property": self._property.to_dict(),
            "start": self._start.isoformat(),
            "end": self._end.isoformat(),
            "extras": self._extras,
            "total_cost": self._total_cost
        }

    @staticmethod
    def from_dict(data, properties):
        tenant = Tenant.from_dict(data["tenant"])
        prop_id = data["property"]["id"]
        prop = next((p for p in properties if p.get_id() == prop_id), None)
        if not prop:
            prop = Property.from_dict(data["property"])
        agr = RentalAgreement(data["id"], tenant, prop, datetime.fromisoformat(data["start"]), datetime.fromisoformat(data["end"]))
        agr._extras = data["extras"]
        agr._total_cost = data["total_cost"]
        return agr

# Декоратор для проверки прав
def check_permissions(role_needed):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            user_role = kwargs.get("role", "guest")
            if user_role != role_needed:
                raise PermissionError(f"Нужна роль {role_needed}, а не {user_role}")
            return func(*args, **kwargs)
        return wrapper
    return decorator

# Фабрика для недвижимости
class PropertyFactory:
    def create(self, type, id, address, area, rate, **kwargs):
        prop_class = PropertyMeta.registry.get(type)
        if not prop_class:
            raise PropertyError(f"Неверный тип недвижимости: {type}")
        return prop_class(id, address, area, rate, **kwargs)

# Цепочка обработки запросов
class RequestHandler:
    def __init__(self):
        self.next = None

    def set_next(self, handler):
        self.next = handler
        return handler

    def handle(self, request, agreement):
        pass

class ManagerHandler(RequestHandler):
    def handle(self, request, agreement):
        if request.get("type") == "minor":
            agreement.log(f"Менеджер одобрил запрос для {agreement._id}")
            return True
        elif self.next:
            return self.next.handle(request, agreement)
        return False

class FinanceHandler(RequestHandler):
    def handle(self, request, agreement):
        if request.get("type") == "cost":
            agreement.log(f"Финансы одобрили запрос для {agreement._id}")
            return True
        elif self.next:
            return self.next.handle(request, agreement)
        return False

class DirectorHandler(RequestHandler):
    def handle(self, request, agreement):
        agreement.log(f"Директор одобрил запрос для {agreement._id}")
        return True

# Процесс аренды
class RentalProcess:
    def rent(self, property, tenant, agent, start, end):
        if not property.is_available():
            raise PropertyError("Недвижимость занята")
        if not self.check_agent(agent):
            raise PermissionError("Нет прав")
        agreement = self.process_rental(property, tenant, start, end)
        property.set_available(False)
        return agreement

    def check_agent(self, agent):
        return False

    def process_rental(self, property, tenant, start, end):
        pass

class OnlineProcess(RentalProcess):
    def check_agent(self, agent):
        return agent.startswith("ONLINE_")

    def process_rental(self, property, tenant, start, end):
        return property.rent_property(tenant, start, end)

class OfflineProcess(RentalProcess):
    def check_agent(self, agent):
        return agent.startswith("OFFLINE_")

    def process_rental(self, property, tenant, start, end):
        return property.rent_property(tenant, start, end)

# Основной сервис
class RentalService:
    def __init__(self):
        self.properties = {}
        self.agreements = {}
        self.handler = ManagerHandler()
        finance = FinanceHandler()
        director = DirectorHandler()
        self.handler.set_next(finance).set_next(director)

    @check_permissions("admin")
    def add_property(self, property, *, role):
        self.properties[property.get_id()] = property
        property.log(f"Добавлена недвижимость {property.get_id()}")

    @check_permissions("admin")
    def remove_property(self, id, *, role):
        if id not in self.properties:
            raise PropertyError("Недвижимость не найдена")
        prop = self.properties.pop(id)
        prop.log(f"Удалена недвижимость {id}")

    def search_address(self, address):
        return [p for p in self.properties.values() if address.lower() in p.get_address().lower()]

    def get_properties(self):
        return list(self.properties.values())

    @check_permissions("manager")
    def handle_request(self, request, agreement_id, *, role):
        if agreement_id not in self.agreements:
            raise AgreementError("Договор не найден")
        return self.handler.handle(request, self.agreements[agreement_id])

    def save(self, file):
        data = {
            "properties": [p.to_dict() for p in self.properties.values()],
            "agreements": [a.to_dict() for a in self.agreements.values()]
        }
        with open(file, 'w') as f:
            json.dump(data, f)
        logging.info(f"Данные сохранены в {file}")

    def load(self, file):
        with open(file, 'r') as f:
            data = json.load(f)
        for p in data["properties"]:
            self.properties[p["id"]] = Property.from_dict(p)
        for a in data["agreements"]:
            self.agreements[a["id"]] = RentalAgreement.from_dict(a, self.get_properties())
        logging.info(f"Данные загружены из {file}")

# Пример
if __name__ == "__main__":
    service = RentalService()
    factory = PropertyFactory()

    # Создаем недвижимость
    try:
        apt = factory.create("apartment", "prop1", "ул. Ленина 5", 60, 800, rooms=2)
        house = factory.create("house", "prop2", "ул. Садовая 10", 120, 1200, garden=True)
        service.add_property(apt, role="admin")
        service.add_property(house, role="admin")

        # Сравнение
        print(apt < house)
        print(apt == house)

        # Аренда
        tenant = Tenant("t1", "Иван Иванов", "ivan@mail.ru")
        process = OnlineProcess()
        agr = process.rent(apt, tenant, "ONLINE_123", datetime(2025, 6, 1), datetime(2026, 6, 1))
        service.agreements[agr._id] = agr
        agr.add_extra("уборка", 50)
        print(agr.report())

        # Запрос
        service.handle_request({"type": "minor"}, agr._id, role="manager")

        # Сохранение и загрузка
        service.save("data.json")
        service.load("data.json")

        # Поиск и отчет
        found = service.search_address("Ленина")
        if found:
            print(found[0])
        print(service.get_properties()[1].report())
    except (PropertyError, PermissionError, AgreementError) as e:
        print(f"Ошибка: {e}")
```

```python
#1
def first(text):

  words = text.split()

  dict = {}

  for word in words:
    if word in dict:
      dict[word] += 1
    else:
      dict[word] = 1

  return dict


text = "book pen square bob red pen book"
print(first(text))
```

```python
#2
def invert_dictionary(original_dict):
    inverted_dict = {}

    for key, value in original_dict.items():
        if value in inverted_dict:
            if isinstance(inverted_dict[value], list):
                inverted_dict[value].append(key)
            else:
                inverted_dict[value] = [inverted_dict[value], key]
        else:
            inverted_dict[value] = key

    return inverted_dict

original = {'a': 1, 'b': 2, 'c': 3}
inverted = invert_dictionary(original)
print(inverted)  # {1: 'a', 2: 'b', 3: 'c'}
```

```python
#3
def merge_dicts(dict1, dict2):
    merged = dict1.copy()

    for key, value in dict2.items():
        if key in merged:
            merged[key] += value
        else:
            merged[key] = value

    return merged

dict1 = {'a': 1, 'b': 2}
dict2 = {'c': 3, 'd': 4}
print(merge_dicts(dict1, dict2))
```

```python
#4
def filter_dict_by_value(dict, num):
    result = {}
    for key, value in dict.items():
        if value > num:
            result[key] = value
    return result

scores = {'Alice': 85, 'Bob': 72, 'Charlie': 90, 'David': 65, 'Eve': 88}

filtered_scores = filter_dict_by_value(scores, 80)
print(filtered_scores)
```

```python
#5
def max_num(dict):
  mx = 0

  for key in dict:
    if dict[key] > mx:
      mx = dict[key]
  return mx

a = {"a": 6, "b": 10, "c": 4, "d": 14, "e": 8}
new_a = max_num(a)
print(new_a)
```

```python
#6
def min_num(dict):
  mn = 1000

  for key in dict:
    if dict[key] < mn:
      mn = dict[key]
  return mn

a = {"a": 6, "b": 10, "c": 4, "d": 14, "e": 8}
new_a = min_num(a)
print(new_a)
```

```python
#7
def remove_keys(dict, keys_to_remove):
    for key in keys_to_remove:
        del dict[key]
    return dict

a = {"a": 6, "b": 10, "c": 4, "d": 14, "e": 8}
b = {"b", "c"}

d = remove_keys(a, b)
print(d)
```

```python
#8
def group_keys_by_length(dict):

    grouped_dict = {}
    for key in dict.keys():
        length = len(key)
        if length in grouped_dict:
            grouped_dict[length].append(key)
        else:
            grouped_dict[length] = [key]
    return grouped_dict


my_dict = {'a': 1, 'bb': 2, 'ccc': 3, 'dd': 4, 'e': 5}
grouped = group_keys_by_length(my_dict)
print(grouped)
```

```python
#9
def grouped_keys(dict, val):

  new_dict = {}

  for key in dict:
    if dict[key] == val:
      new_dict[key] = dict[key]
  return new_dict


a = {"a": 6, "b": 10, "c": 10, "d": 10, "e": 8}

new = grouped_keys(a, 10)
print(new)
```

```python
#10

def grouped_keys(dict1, dict2):

  new_dict = {}

  for key in dict1:
    new_dict[key] = dict1[key] + dict2[key]
  return new_dict

a = {"a": 6, "b": 10, "c": 10, "d": 10, "e": 8}
b = {"a": 2, "b": 7, "c": 4, "d": 14, "e": 5}

new = grouped_keys(a,b)
print(new)
```

```python
#4
def first(tup, elem):
  index = 0
  n = len(tup)
  for i in range(n):
    if tup[i] == elem:
      index = i
  return index

student = ("bob", 19, "KFU", "Kazan")
print(first(student, 19))
```

```python
#5
def count_el(tup, elem):
  k = 0
  n = len(tup)
  for i in range(n):
    if tup[i] == elem:
      k += 1
  return k

student = ("bob", 19, "KFU", "Kazan", 19, 19)
print(count_el(student, 19))
```

```python
#8
def sort_tuple(input_tuple):
    lst = list(input_tuple)
    n = len(lst)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lst[j] > lst[j + 1]:
                lst[j], lst[j + 1] = lst[j + 1], lst[j]
    return lst

numbers = (5, 2, 8, 1, 3)
print(sort_tuple(numbers))
```

```python
#9
def two(tup):
  n = len(tup) // 2
  lst1 = tup[:n]
  lst2 = tup[n:]
  lst3 = [lst1, lst2]
  return lst3

numbers = (7, 5, 2, 8, 1, 3)
print(two(numbers))
```

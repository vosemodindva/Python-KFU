```python
x = []
y = []

n = int(input())

for i in range(n):
  x.append(int(input()))

for i in range(1, n-1):
  if (x[i-1] < x[i] > x[i+1]):
    y.append(x[i])
print(y)
print(f'максимальный из локальных максимумов = {max(y)}')
print(x)
```

```python
s = "HEL:LO :WOR:LD:"
n = len(s)
k = 0
for i in range(n):
  if s[i] == ":":
    k += 1
new = s.replace(':','')
print(new, f', count of ":" = {k}')
```

```python
s = "world, book, pen, pillar, window, blow."
k = 0
new = s.split()
for n in new:
  if n[-2] == 'w':
    k+=1
print(f'кол-во слов с "W" на конце {k}')
```

```python
def create_file(file_name):
    with open(file_name, "w") as file:
        file.write("abc def ghi jkl")

if __name__ == "__main__":
    file_name = "ex1.txt"
    create_file(file_name)

    with open(file_name, "r") as file:
        content = file.read()


    last_space_index = content.rfind(' ')

    if last_space_index != -1:

        new_content = content[:last_space_index]

        print("Новая строка:", new_content)

        with open(file_name, "w") as file:
            file.write(new_content)
    else:
        print("В файле нет пробелов")
```

```python
def create_file(file_name):
  with open(file_name, "w") as file:
    file.write("4 1 3 2 5 9 6")

if __name__ == "__main__":
  file_name = "ex2.txt"
  create_file(file_name)

  with open('ex2.txt', 'r') as file:
    s = list(map(int, file.read().split()))

  mx = max(s)
  max_index = s.index(mx)
  mn = min(s)
  min_index = s.index(mn)

  s[max_index],s[min_index] = s[min_index],s[max_index]

  with open("ex2.txt", "r") as file:
    for line in file:
      print(f'старый файл: {line}')

  with open("new_ex2.txt", "w") as file:
    file.write(' '.join(map(str, s)))

  with open("new_ex2.txt", "r") as file:
    for line in file:
      print(f'новый файл(max и min поменяли): {line}')
```

```python
def create_file(file_name):
  with open(file_name, "w") as file:
    file.write("1 2 3 4 5 6")

if __name__ == "__main__":
  file_name = "input.txt"
  create_file(file_name)

  with open('input.txt', 'r') as file:
    numbers = list(map(int, file.read().split()))

# Обратный порядок
  reversed_numbers = numbers[::-1]

# Запись в новый файл
  with open('output.txt', 'w') as file:
    file.write(' '.join(map(str, reversed_numbers)))

  with open("output.txt", "r") as file:
    for line in file:
      print(line)
```
